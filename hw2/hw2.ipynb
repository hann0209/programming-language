```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
data = pd.read_csv('prices.csv')

# Filter data for male and female
male_data = data[data['Gender'] == 'male']
female_data = data[data['Gender'] == 'female']

# Calculate average salary for male and female
avg_male_salary = male_data['Salary'].mean()
avg_female_salary = female_data['Salary'].mean()

# bar chart
plt.bar(['male', 'female'], [avg_male_salary, avg_female_salary], color=['blue', 'pink'])
plt.title('Average Salary Comparison between Male and Female Employees')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Salary'], color='skyblue', alpha=0.6)
plt.title('Salary Trend by Age')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
age_group_counts = data['Age'].value_counts().sort_index()

# the distribution of employee age groups
plt.figure(figsize=(10, 6))
plt.plot(age_group_counts.index, age_group_counts.values, marker='o', color='skyblue', linestyle='-')
plt.title('Employee Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Highlighting the age group with the highest employment
most_common_age_group = age_group_counts.idxmax()
plt.axvline(x=most_common_age_group, color='red', linestyle='--', label='Most Common Age Group')
plt.legend()

plt.show()
```
